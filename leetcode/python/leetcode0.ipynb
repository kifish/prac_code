```python
class Solution:
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        idx2val = {}
        for idx,item in enumerate(nums):
            idx2val[idx] = item
        L = sorted(idx2val.items(),key = lambda x:x[1])  
        r_L = sorted(idx2val.items(),key = lambda x:x[1],reverse = True)
        
        for k,v in L:
            temp = target - v
            for r_k,v in r_L:
                if v == temp and k != r_k:
                    return [k,r_k]
```

```python
class Solution:
    def reverse(self, x):
        """
        :type x: int
        :rtype: int
        """
        if self.judge(x):  
            x = str(x)
            if x[0] == '-':
                temp = int(x[:0:-1]) 
                if self.judge(temp):
                    return -1*int(x[:0:-1])
                else:
                    return 0
            else:
                temp = int(x[::-1])
                if self.judge(temp):
                    return temp
                else:
                    return 0 
        else:
            return 0
        
    def judge(self,x):
        if x > pow(2,31) -1 or x < -1*pow(2,31):
            return False
        return True
```

Of course, the solution is still O(n^2) in the worst case, but it can be faster than most implements under given test cases.
Two key points are: (1) a quick check see if the new value is already the largest (2) only refresh the search pointer p when the target is before it, in other words smaller.

```python
# Definition for singly-linked list.

class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None
# add sentinel node
class Solution:
    def testcase(self,x):
        if len(x) > 0:
            head = ListNode(x[0])
            cur = head
        for i in range(1,len(x)):
            new_node = ListNode(x[i])
            cur.next = new_node
            cur = cur.next
        return head
    
    def insertionSortList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        """
        # 根据夹逼，确定要插入的位置
        sentinel_node = ListNode(0)
        sentinel_node.next = head
        sorted_tail = head
        pre_to_add = sentinel_node
        
        while sorted_tail and sorted_tail.next:
            next_node = sorted_tail.next
            if next_node.val >= sorted_tail.val:
                sorted_tail = sorted_tail.next
                continue
            
            if pre_to_add.next.val > next_node.val:
                pre_to_add = sentinel_node
            
            #遍历的起点的下一个节点的值必须保证小于等于要插入的值
            #不能保证(如果已经大于了，那么可能已经过了站)，则从头结点开始重新搜
            # 第一个出现 比 插入值 大的值的位置 即是要插入的位置的附近
            #这样的话查找位置就是O(n)
            while pre_to_add.next.val <= next_node.val:
                pre_to_add = pre_to_add.next
            #处理尾部，为下次循环做准备
            sorted_tail.next = next_node.next
            #处理插入
            next_node.next = pre_to_add.next
            pre_to_add.next = next_node
        return sentinel_node.next
            
                
                
    def print_(self,head):
        if head:
            cur = head
            while cur:
                print(cur.val)
                cur = cur.next
     
```

```python
print(len(x))
a = Solution()
import time
s = time.time()
a.insertionSortList(a.testcase(x))
print(time.time() - s)
```

```python
class Solution:
    def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        val_unique = set()
        for num in nums:
            if num in val_unique:
                nums.remove(num)
            else:
                val_unique.add(num)
        return len(nums)
    def __init__(self,nums):
        self.nums = nums
```

```python
class Solution:
    def myAtoi(self, str):
        """
        :type str: str
        :rtype: int
        """
        max_int = 2**31-1
        min_int = -1*(2**31)
        is_num = False
        sign = ''
        num_str = ''
        for char in str:
            if char == ' ':
                if is_num:
                    break
                else:
                    if sign != '':
                        is_num = False
                        break
                    else:
                        continue
            if char == '+' or char == '-':
                if sign != '':
                    break
                else:
                    if is_num:
                        break
                    else:
                        sign = char
                        continue
                
            if self.valid_char(char):
                if '0' <= char <= '9':
                    is_num = True 
                    num_str += char
                    continue
            
            if char == ' ' or (char > '9' or char < '0'):
                break
        if is_num:
            num_str = sign + num_str
            num = int(num_str)
            num = num if num <= max_int else max_int
            num = num if num >= min_int else min_int
            return num
        else:
            return 0
        
    def valid_char(self,input):
        if input[0] != ' ' and ('A' <= input[0] <= 'z' or '0' <= input[0] <= '9'):
            return True
        else:
            return False
    
    def testcase(self,input):
        return self.myAtoi(input)
        
```
