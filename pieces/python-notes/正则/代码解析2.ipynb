```python
def getSampleNo(text):
    digit_alph_no=[]
    regex_digit_alph = '\d{2}S\d{5}\-\d{3}|M\d{4}[\-~]\d{4}|\d{4}\-[A-Z]\d{4,5}'
    sn_res=re.findall(regex_digit_alph, text)
    for s in sn_res:
        digit_alph_no.append(s)
    text = re.sub(regex_digit_alph, DIGIT_ALPHA, text)
    return text, digit_alph_no
```
