如果有global 则形参不能和global变量ambiguous

```
x = 1
def change(x):
    global x
    x = 2

change(x)
print(x)
```

```
x = 1
def change():
    global x
    x = 2

change()
print(x)
```

```
def foo(a):
   """在函数内部直接修改了同一个引用指向的对象。
   也就修改了实际参数传来的引用值指向的对象。
   """
   a.append("can change object")
   return a

lst = [1,2,3]
print(foo(lst))
print(lst)
```

```
def foo(a):
   """
   实际参数传来一个对象[1,2,3]的引用.形式参数（局部变量a
   重新绑定到新的对象，当然不能修改原来的对象了。
   """
   a = ["Python","test"]
   return a

lst = [1,2,3]
for item in foo(lst):
   print(item)
print(lst)
```

要理解python 中的 = 其实是binding


"copy" is misleading. Python doesn't have variables like C. All names in Python are references. You can't modify name, you just can bind it to another object, that's all. It only makes sense to talk about mutable and immutable object in Python not they are names

```
def f(n, x):  # these `n`, `x` have nothing to do with `n` and `x` from main()
    n = 2 # put `n` label on `2` balloon
    x.append(4)  # call `append` method of whatever object `x` is referring to.
    print('In f():', n, x) 

    
def main():
    n = 1
    x = [0,1,2,3]
    print('Before:', n, x)
    f(n, x)
    print('After: ', n, x)

main()
```

```
def f(n, x):  # these `n`, `x` have nothing to do with `n` and `x` from main()
    n = 2 # put `n` label on `2` balloon
    x.append(4)  # call `append` method of whatever object `x` is referring to.
    print('In f():', n, x) 
    x = []   # put `x` label on `[]` ballon
    # x = [] has no effect on the original list that is passed into the function
    

def main():
    n = 1
    x = [0,1,2,3]
    print('Before:', n, x)
    f(n, x)
    print('After: ', n, x)

main()
```

传参进来 直接就是name 来引用, 如果是 name = xxxx 那么会重新绑定, 如果对引用的对象进行修改,那么可以对原对象生效, 如果是 name = new value , 那么实际上 只是重新绑定,影响不了原值.
