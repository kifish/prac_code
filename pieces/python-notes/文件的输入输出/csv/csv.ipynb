```
# -*- coding: UTF-8 -*-


import csv


def read_data(filename):
    #data = {}
    new_data = {}
    with open(filename) as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            ID = row["ID"]
            del row["ID"]
            #data[ID] = row
            new_data[ID] = dict(list(row.items()))
    return new_data

def clean_data(data):
    cleaned_dict = {}             #initialize
    cnt = 0;
    for key,val in data.items():
        cur_part = {}                   #initialize
        islegal = True
        for cur_key, cur_val in val.items():         #loop
            if cur_val == 'NA':          #modify
                islegal = False
                break
            elif cur_key=='DebtRatio'and cur_val > 1:
                islegal = False
                break
            else:
                cur_part[cur_key] = cur_val
        if islegal:
            cleaned_dict[cnt] = cur_part
            cnt += 1
    return cleaned_dict,cnt

filename = 'C:\\Users\\Administrator\\Desktop\\cs-training2.csv'
data = read_data(filename)
print(data)
cleaned_data ,num = clean_data(data)
print(num)
cnt = 0;
#print(cleaned_data)
for key in cleaned_data.keys():
    if cnt < 10:
        print(key,cleaned_data[key])
    cnt+= 1

'''
from sklearn.linear_model import LinearRegression
train = data[data.MonthlyIncome.isnull() == False]
X_train = train.drop(['MonthlyIncome', 'SeriousDlqin2yrs'], axis=1)
y_train = train.MonthlyIncome
lmMod = LinearRegression(fit_intercept=True, normalize=True).fit(X_train, y_train)

'''
```
