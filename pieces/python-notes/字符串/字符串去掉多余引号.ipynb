去掉多余引号

```
    while True:
        cut_res = input('输入:')  # "肝组织", "1个", ",", "大小", "16×10×8CM"
        cut_res = cut_res.split(',') # 这里有点问题,这里把所有的逗号都去掉了
        print(cut_res)

        print()
        new_cut_res = []
        for item in cut_res:
            new_item = ""
            for char in item:
                if char != '\'' and char != '\"':
                    if char != ' ':
                        new_item += char
            if len(new_item) > 0:
                new_cut_res.append(new_item)

        cut_res = new_cut_res
        print(cut_res)
```
