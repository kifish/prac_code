```
def is_odd(n):
    return n % 2 == 1

#如果返回结果为空呢
print(filter(is_odd, [2,4,6]))
print('---------')
#如果转list了，返回空list，如果没转，相当于一个generator
print(list(filter(is_odd, [2,4,6])))
print('---------')
res = list(filter(is_odd, [2,4,6]))
print(res)
```

```
ori_list = ['江苏',' ','','南京']
print(ori_list)
```

```
condition = lambda x : x != '' and x != ' '
print(list(filter(condition, ori_list)))
```
